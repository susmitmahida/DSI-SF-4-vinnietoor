```python
headers = {
    1: "Player",
    2: "Team",
    3: "GP",
    4: "MIN", 
    5: "FGM",
    6: "FGA",
    7: "FG%",
    8: "3PM",
    9: "3PA",
    10: "3P%",
    11: "FTM",
    12: "FTA",
    13: "FT%",
    14: "TOV",
    15: "PF",
    16: "ORB",
    17: "DRB",
    18: "REB",
    19: "AST",
    20: "STL",
    21: "BLK",
    22: "PTS"
    
    
}

row_data = []

table_rows = Selector(text=driver.page_source).xpath("//table/tbody/tr")

for row in table_rows:
    tds       =  row.xpath("./td")
    td_data   =  {}
    for index, col in enumerate(tds):
        if index > len(headers) or index == 0:
            continue
        soup       =  BeautifulSoup(col.extract())
        col_value  =  soup.text
        td_data.update({headers[index]:col_value})
    row_data.append(td_data)
    
    
row_data
cs1 = pd.DataFrame(row_data)
cs1
```
